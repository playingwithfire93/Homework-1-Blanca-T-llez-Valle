# Question 1


### **Create a pie chart which shows the proportion of male / female passengers.**

```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
gender_counts = titanic["sex"].value_counts()
```

```python
# Importing necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn  # Note: sklearn is imported but not used in this code snippet

# Setting the figure size for the pie chart
plt.figure(figsize=(6, 6))

# Creating a pie chart with gender_counts data
# Labels represent the unique values (Male and Female)
# Autopct formats the percentage display on the chart
# Startangle determines the starting angle of the pie chart (90 degrees in this case)
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)

# Adding a title to the pie chart
plt.title('Proportion of Male/Female Passengers')

# Displaying the pie chart
plt.show()
```

```python
# Count the number of passengers in each class and store the counts in a variable.
class_counts = titanic['class'].value_counts()

# Plot a pie chart for class distribution
plt.figure(figsize=(6, 6))  # Set the size of the figure (width, height) in inches.

# Create a pie chart using the class_counts data, with labels and percentage format, starting the angle at 90 degrees.
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)

# Add a title to the pie chart.
plt.title('Proportion of Passengers in Each Class')

# Display the pie chart.
plt.show()
```

# Question 3

### **Explain the relationship between age and fare using a visualization.**

```python
# Importing the necessary library for plotting
import matplotlib.pyplot as plt

# Creating a figure with a specified size (8 inches by 6 inches)
plt.figure(figsize=(8, 6))

# Creating a scatter plot using age as the x-axis and fare as the y-axis
# Setting alpha to 0.5 for transparency in data points
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5)

# Adding a title to the plot
plt.title('Relationship between Age and Fare')

# Labeling the x-axis with 'Age'
plt.xlabel('Age')

# Labeling the y-axis with 'Fare'
plt.ylabel('Fare')

# Displaying the plot
plt.show()
```

#### The figure suggests that there is no distinct relationship between age and fare, although we observe an increase in fares within the age range of 20-60, with some outliers in the 30-40 range. Due to the large amount of data and the limitations of the plot, it is necessary to calculate the correlation coefficients to determine if there is a clear correlation between age and fare.

```python
# Selecting columns 'age' and 'fare' from the Titanic dataset and creating a new DataFrame called 'titanic_age_fare'
titanic_age_fare = titanic[['age', 'fare']]

# Calculating the correlation matrix for the selected columns
correlation = titanic_age_fare.corr()

# Generating a heatmap to visualize the correlation matrix, with annotations and a coolwarm color map
sns.heatmap(correlation, annot=True, cmap='coolwarm')

# Adding a title to the heatmap
plt.title('Correlation between Age and Fare')

# Displaying the heatmap
plt.show()
```

#### This indicates a very weak positive correlation between age and fare. In other words, there is a slight tendency for the fare to increase slightly as age increases, but the relationship is not strong. 

# Question 4

### **What is the survival ratio by age, sex, and class.**

```python
# Binning Age into groups
age_bins = [0, 18, 30, 50, 100]
age_labels = ['0-18', '19-30', '31-50', '51+']
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=age_bins, labels=age_labels)

# Create a bar plot for survival ratio by age, sex, and class
plt.figure(figsize=(14, 8))  # Set the figure size

# Survival ratio by Age
plt.subplot(2, 2, 1)  # Create a subplot in a 2x2 grid at position 1
sns.barplot(x='AgeGroup', y='survived', data=titanic, errorbar=None)  # Bar plot for survival ratio by age
plt.title('Survival Ratio by Age Group')  # Set the subplot title
plt.xlabel('Age Group')  # Set the x-axis label
plt.ylabel('Survival Ratio')  # Set the y-axis label

# Survival ratio by Sex
plt.subplot(2, 2, 2)  # Create a subplot in a 2x2 grid at position 2
sns.barplot(x='sex', y='survived', data=titanic, errorbar=None)  # Bar plot for survival ratio by sex
plt.title('Survival Ratio by Sex')  # Set the subplot title
plt.xlabel('Sex')  # Set the x-axis label
plt.ylabel('Survival Ratio')  # Set the y-axis label

# Survival ratio by Class
plt.subplot(2, 2, 3)  # Create a subplot in a 2x2 grid at position 3
sns.barplot(x='class', y='survived', data=titanic, errorbar=None)  # Bar plot for survival ratio by class
plt.title('Survival Ratio by Class')  # Set the subplot title
plt.xlabel('Class')  # Set the x-axis label
plt.ylabel('Survival Ratio')  # Set the y-axis label

# Survival ratio by Age and Sex with hue for Class
plt.subplot(2, 2, 4)  # Create a subplot in a 2x2 grid at position 4
sns.barplot(x='AgeGroup', y='survived', hue='sex', data=titanic, errorbar=None)  # Bar plot for survival ratio by age and sex with class as hue
plt.title('Survival Ratio by Age Group and Sex')  # Set the subplot title
plt.xlabel('Age Group')  # Set the x-axis label
plt.ylabel('Survival Ratio')  # Set the y-axis label
plt.legend(title='Sex', loc='upper right')  # Add legend with title and position

# Adjust layout
plt.tight_layout()  # Adjust layout for better visualization
plt.show()  # Display the plot
```

#### The data reveals a significantly higher survival ratio for females compared to males, and this is plausibly explained by adherence to social norms of the time, which prioritized "women and children first" during the Titanic disaster.


#### Moreover, the data indicates a notably higher survival ratio for first-class passengers compared to other classes. This can be attributed to several factors, including the limited number of lifeboats that were prioritized for first-class individuals. Additionally, the proximity of first and second-class cabins to lifeboat stations facilitated quicker access for these passengers, while those in third class, often located in lower decks, faced greater distance and obstacles in reaching the lifeboats. The advantage in receiving timely information about the severity of the situation also favored first and second-class passengers, allowing them more time to respond and evacuate.

#### Analyzing age-related survival ratios, the group with the highest survival rate is 0-18, consistent with the social norm of prioritizing children and women during evacuation. Another age group showing high survival is 30-50. This aligns with the previous observation that there is a weak correlation between age and fare, and a scatterplot suggests that higher fares are associated with individuals aged 30-50, possibly indicating first or second-class passengers.
