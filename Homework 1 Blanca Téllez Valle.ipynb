#### What is the difference between a Pandas DataFrame and a Pandas series. Show an example of how you create each of them.

##### **Pandas DataFrame:**

A Pandas DataFrame is a labeled data structure that is two dimensional and has columns that can hold various types of data. 
Imagine it as a table or spreadsheet with Pandas Series in each column. In particular, DataFrames are helpful for working with structured data.


###### **Example of Pandas Dataframe**

```python
import pandas as pd
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Blanca', 'Mats', 'Eun-Sik', 'Amandla'],
    'Age': [30, 57, 35, 20],
    'City': ['Málaga', 'Gothenburg', 'Seoul', 'Cape Town']
}
df = pd.DataFrame(data)
print(df)
```

In this example, a dictionary is used to generate a Pandas DataFrame, with keys indicating column names and values, lists that represent the data in each column. A tabular structure with labeled rows and columns is called a DataFrame. 


##### **Pandas Series:**

A Pandas Series is essentially a one-dimensional labeled array. Every element in the Series is given a label known as an index, and it can hold any kind of data.


###### **Example of Pandas Series**

```python
import pandas as pd
# Create a Series from a list
data = [1, 3, 5, 7, 9]
series = pd.Series(data)
print(series)
```

In this example, a Python list is used to construct a Pandas Series. The data type of the Series is deduced from the data, and the index is automatically assigned.

##### **Differences:**

- A data frame has two dimensions, whereas a series has one.
- In a data frame, the data types of the elements do not have to match, but in a series, every element needs to have the same data type.
- Data frames have row and column indices, whereas series only have a single index. 
- Data frames, on the other hand, may manage data with several variables, while series are utilized when there is just one variable or column.

##### Give an overview (code and an explanation) of all missing values in the data. 

```python
missing_values = titanic.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

The **isnull()** method is used to return a DataFrame where each entry is True if the corresponding data point is null and False if there is data. 

The subsequent **.sum()** method is then used to calculate the sum of True values along each column and counting them.

The result is a Series named **missing_values** that provides the count of missing values for each column in the dataset.

Finally, the code prints out the statement "Missing values per column:" followed by the missing_values Series, offering an overview of the extent of missing data in the Titanic dataset.

In this case there's only missing data in 4 categories:
- "age" with 177 missing values.
- "embark" and "embark_town" with 2 missing values each.
- "deck" with 688 missing values. 



##### Delete the rows where you do not have information about the age of the person. Then group the passengers in a 10 year age range (for example, you can do something like 0 – 10, 11 – 20, 21 – 30, etc). 


The method .dropna() is used to eliminate all of the rows that have missing data in the age category.

```python
titanic_cleaned = titanic.dropna(subset=['age'])
```

The ".cut() method is like putting passengers into groups based on their ages, with each group covering a range of 10 years. Since there aren't any passengers older than 90, the last group is made for those between 80 and 90 years old."

```python
# Assuming 'titanic_cleaned' is your DataFrame
age_categories = pd.cut(titanic_cleaned['age'], bins=range(0, 91, 10), right=False)
titanic_cleaned.loc[:, 'Age_Category'] = age_categories
```

##### For each age category created in d), find out how many passengers are female/male, and how many travelled in each class. 

```python
age_category_stats = titanic_cleaned.groupby(['Age_Category', 'sex', 'pclass'], observed=False).size().reset_index(name='Passenger_Count')
print(age_category_stats)
```
